# Covid Detaction 

```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/COVID19%20data%20for%20overall%20INDIA.csv")
data.head()
```

```python
data.info
```

```python
data.isnull().sum()
```

```python
import plotly.express as px

fig1 = px.histogram(data, x=data['Date_YMD'], y=data['Daily Confirmed'])
fig2 = px.histogram(data, x=data['Date_YMD'], y=data['Daily Deceased'])
fig1.show()
fig2.show()
```

```python
cases = data['Daily Confirmed'].sum()
deceased = data['Daily Deceased'].sum()

import plotly.express as px

fig3 = px.pie(data,
              values=[cases, deceased], 
              names=['Comfirm', 'Deceased'],
               title='Daily Confirmed Cases vs Daily Deaths', hole=0.5)
fig3.show()
```

###  Covid-19 Death Prediction

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(data['Daily Deceased'], order=(2, 1, 2))
model_fit = model.fit()
forecast = model_fit.forecast(steps=7)  # next 7 days

print(forecast)
```
